```python
# Unzip Task 2 data - CHáº¬M, tá»«ng file má»™t Ä‘á»ƒ khÃ´ng spam API
import zipfile
import time
import os

def unzip_slowly(zip_path, extract_to, delay=0.001):
    """Giáº£i nÃ©n cháº­m Ä‘á»ƒ khÃ´ng spam Drive API"""
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        members = zip_ref.namelist()
        total = len(members)
        
        for i, member in enumerate(members):
            zip_ref.extract(member, extract_to)
            if i % 100 == 0:
                print(f"Progress: {i}/{total} ({i*100//total}%)")
                time.sleep(0.1)  # Nghá»‰ má»—i 100 files
        
        print(f"âœ… Done: {total} files")

print("Unzipping Task 2 train...")
unzip_slowly(
    '/content/drive/MyDrive/THESIS/data/task2/train.zip',
    '/content/drive/MyDrive/THESIS/data/task2/'
)

print("\nUnzipping Task 2 val...")
unzip_slowly(
    '/content/drive/MyDrive/THESIS/data/task2/val.zip',
    '/content/drive/MyDrive/THESIS/data/task2/'
)
```

```python
# Unzip Task 3 data
print("Unzipping Task 3 train...")
unzip_slowly(
    '/content/drive/MyDrive/THESIS/data/task3/train.zip',
    '/content/drive/MyDrive/THESIS/data/task3/'
)

print("\nUnzipping Task 3 val...")
unzip_slowly(
    '/content/drive/MyDrive/THESIS/data/task3/val.zip',
    '/content/drive/MyDrive/THESIS/data/task3/'
)
```
