```python
# Unzip Task 2 data - NHANH trÃªn Kaggle (khÃ´ng cáº§n delay)
import zipfile
import os

def unzip_data(zip_path, extract_to):
    """Giáº£i nÃ©n data"""
    os.makedirs(extract_to, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        members = zip_ref.namelist()
        total = len(members)
        
        for i, member in enumerate(members):
            zip_ref.extract(member, extract_to)
            if i % 100 == 0:
                print(f"Progress: {i}/{total} ({i*100//total}%)")
        
        print(f"âœ… Done: {total} files")

print("Unzipping Task 2 train...")
unzip_data(
    '/kaggle/input/thesis-data/task2/train.zip',
    '/kaggle/working/data/task2/'
)

print("\nUnzipping Task 2 val...")
unzip_data(
    '/kaggle/input/thesis-data/task2/val.zip',
    '/kaggle/working/data/task2/'
)
```

```python
# Unzip Task 3 data
print("Unzipping Task 3 train...")
unzip_data(
    '/kaggle/input/thesis-data/task3/train.zip',
    '/kaggle/working/data/task3/'
)

print("\nUnzipping Task 3 val...")
unzip_data(
    '/kaggle/input/thesis-data/task3/val.zip',
    '/kaggle/working/data/task3/'
)
```
