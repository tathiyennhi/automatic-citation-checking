## Tại sao không cần?

Khi bạn upload file `.zip` lên Kaggle Dataset:
- Kaggle tự động giải nén
- Data đã có sẵn ở `/kaggle/input/thesis-data-task2/train/train/`
- **KHÔNG CẦN unzip thêm nữa!**

## Bạn cần làm gì?

1. Upload 2 datasets:
   - `thesis-data-task2` (chứa train.zip, val.zip)
   - `thesis-data-task3` (chứa train.zip, val.zip)

2. Chạy trực tiếp training notebooks:
   - `Task2_BERT_Training.ipynb`
   - `Task2_SciBERT_Training.ipynb`
   - `Task2_RoBERTa_Training.ipynb`
   - `Task3_BERT_Training.ipynb`
   - `Task3_SciBERT_Training.ipynb`
   - `Task3_RoBERTa_Training.ipynb`

3. **XONG!** Không cần unzip gì cả!
